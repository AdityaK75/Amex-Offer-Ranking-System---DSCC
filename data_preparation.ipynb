```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_parquet("/Users/adityakanagalekar/Downloads/train_data-2.parquet")
test_df = pd.read_parquet("/Users/adityakanagalekar/Downloads/test_data-2.parquet")
trans_df = pd.read_parquet("/Users/adityakanagalekar/Downloads/add_trans.parquet")
event_df = pd.read_parquet("/Users/adityakanagalekar/Downloads/add_event.parquet")
offer_metadata_df = pd.read_parquet("/Users/adityakanagalekar/Downloads/offer_metadata.parquet")

print(train_df.head())
print(test_df.head())
print(trans_df.head())
print(event_df.head())
print(offer_metadata_df.head())
```

```python
train_df
```

```python
for i in train_df:
    print(i, train_df[i].nunique(), train_df[i].dtype)
```

```python
# List columns with dtype 'object'
object_columns = [col for col in train_df.columns if train_df[col].dtype == 'object']
print("Columns with dtype 'object':", object_columns)
```

```python
# Convert 'id4' and 'id5' columns to datetime
train_df['id4'] = pd.to_datetime(train_df['id4'], errors='coerce')
train_df['id5'] = pd.to_datetime(train_df['id5'], errors='coerce')
test_df['id4'] = pd.to_datetime(test_df['id4'], errors='coerce')
test_df['id5'] = pd.to_datetime(test_df['id5'], errors='coerce')
```

```python
train_df.isna().sum()
```

```python
for i in offer_metadata_df:
    print(i, offer_metadata_df[i].nunique(), offer_metadata_df[i].dtype)
```

```python
# Drop specified columns from offer_metadata_df
offer_metadata_df = offer_metadata_df.drop(columns=['f377', 'id9', 'id11'])
# Replace all values in 'f378' that are not 'N' with 'Y'
offer_metadata_df['f378'] = offer_metadata_df['f378'].apply(lambda x: 'N' if x == 'N' else 'Y')
```

```python
offer_metadata_df['id12'] = pd.to_datetime(offer_metadata_df['id12'], errors='coerce')
offer_metadata_df['id13'] = pd.to_datetime(offer_metadata_df['id13'], errors='coerce')
```

```python
offer_metadata_df
```

```python
# Merge train_df with offer_metadata_df on 'id3' (offer_id)
train_merged_df = pd.merge(train_df, offer_metadata_df, how='left', on='id3')
test_merged_df = pd.merge(test_df, offer_metadata_df, how='left', on='id3')

# Display the merged DataFrame
train_merged_df.head()
```

```python
train_merged_df['id12'].isna().sum()
```

```python
# Drop rows where 'id12' is NA from train_merged_df
train_merged_df = train_merged_df.dropna(subset=['id12'])
test_merged_df = test_merged_df.dropna(subset=['id12'])

# Show the new shape to confirm rows were dropped
print(train_merged_df.shape)
print(test_merged_df.shape)
```

```python
for i in event_df:
    print(i, event_df[i].nunique(), event_df[i].dtype)
```

```python
event_df['id4'] = pd.to_datetime(event_df['id4'], errors='coerce')
event_df['id7'] = pd.to_datetime(event_df['id7'], errors='coerce')
```

```python
# Merge test_merged_df with event_df on columns 'id2', 'id3', and 'id4'
test_event_merged_df = pd.merge(test_merged_df, event_df, how='left', on=['id2', 'id3', 'id4'])

test_event_merged_df.head()
```

```python
train_event_merged_df = pd.merge(train_merged_df, event_df, how='left', on=['id2', 'id3', 'id4'])   
```

```python
for i in trans_df:
    print(i, trans_df[i].nunique(), trans_df[i].dtype)
```

```python
trans_df['f370'] = pd.to_datetime(trans_df['f370'], errors='coerce')
trans_df['f371'] = pd.to_datetime(trans_df['f371'], errors='coerce')
```

```python
# Ensure id7 and f370 are datetime
train_event_merged_df['id7'] = pd.to_datetime(train_event_merged_df['id7'], errors='coerce')
test_event_merged_df['id7'] = pd.to_datetime(test_event_merged_df['id7'], errors='coerce')

trans_df['f370'] = pd.to_datetime(trans_df['f370'], errors='coerce')

# Extract date part (without time) and convert to datetime64[ns]
train_event_merged_df['event_date'] = pd.to_datetime(train_event_merged_df['id7'].dt.date)
test_event_merged_df['event_date'] = pd.to_datetime(test_event_merged_df['id7'].dt.date)

trans_df['trans_date'] = pd.to_datetime(trans_df['f370'].dt.date)

# Merge on id2, id8, and date
train_trans_merged_df = pd.merge(
    train_event_merged_df,
    trans_df,
    how='left',
    left_on=['id2', 'id8', 'event_date'],
    right_on=['id2', 'id8', 'trans_date']
)

test_trans_merged_df = pd.merge(
    test_event_merged_df,
    trans_df,
    how='left',
    left_on=['id2', 'id8', 'event_date'],
    right_on=['id2', 'id8', 'trans_date']
)

train_trans_merged_df.head()
```
